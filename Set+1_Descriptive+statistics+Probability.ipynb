```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x=pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
x
```

```python
Company=['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
#pie chart
sns.set_palette('pastel')
plt.figure(figsize = (8, 6))
plt.pie(x, labels = Company, autopct = '%1.0f%%')
plt.title('Companies', fontsize = 18, fontweight = 'bold')
plt.show()
```

```python
# Box Plot to find outliars
sns.boxplot(x)
```

#### Inference: ####
- From the box plot we can observe one outlier at 91.36% which is Morgan Stanley.
